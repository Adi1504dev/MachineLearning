### Custom Exception

```python
class Error(Exception):
    pass
#Here we create error and then inherit to it but we could have
# done direclty by inheriting Exception , we have done this so Error
# acts as generic class for custom exception
class dobException(Error):
    """Raised when the date of birth is invalid"""
    def __init__(self, message="Date of Birth is invalid. Please enter in DD-MM-YYYY format"):
        self.message = message
        super().__init__(self.message)
class emailException(Error):
    """Raised when the email is invalid"""
    def __init__(self, message="Email is invalid. Please enter a valid email address"):
        self.message = message
        super().__init__(self.message)
```

```python
age=35
try:
    if age<20 and age<30:
        print("Eligible")
    else:
        raise dobException
except dobException:
    print(dobException.message)
```
