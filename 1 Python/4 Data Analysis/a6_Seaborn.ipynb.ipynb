### Seaborn
Seaborn in python visualization library based on mathplot lib that provides high level interface for drawing attractive and informative statical graphics.  It helps in creating complex visualization with just a few line of code . In this lessom we will cover the basics of seaborn , including creating various types of plots and customizing them

#### Basic plotting

```python
import seaborn as sns
#there are many internal data set in it
tips=sns.load_dataset("tips")
tips
```

#### Create scatterplot


```python
import matplotlib.pyplot as plt
sns.scatterplot(x='total_bill',y='tip',data=tips)
plt.title("Total Bill vs tip")
```

#### Line plot

```python
import matplotlib.pyplot as plt
sns.lineplot(x='size',y='total_bill',data=tips)
plt.title("Total Bill by size")
```

#### Categorical plot

##### Bar Plot


```python
import matplotlib.pyplot as plt
sns.barplot(x='day',y='total_bill',data=tips)
plt.title("Total Bill by day")
```

##### Box Plot



```python
import matplotlib.pyplot as plt
sns.boxplot(x='day',y='total_bill',data=tips)
plt.title("Total Bill by day")
```

##### Violin plot

```python
import matplotlib.pyplot as plt
sns.violinplot(x='day',y='total_bill',data=tips)
plt.title("Total Bill by day")
```

##### Histogram


```python
import matplotlib.pyplot as plt
sns.histplot(x=tips['total_bill'],bins=10 ,kde=True)
#kde true give line on histogram defining its shape
plt.title("Bill histogram")
```

##### KDE plot

```python
import matplotlib.pyplot as plt
sns.kdeplot(x=tips['total_bill'] ,fill=True)
#kde true give line on histogram defining its shape
plt.title("Bill histogram")
```

#### PairPlot


```python
sns.pairplot(tips)
#It generates pair plot between all numerical values in data set
```

#### Heatmap

```python
#While selecting cloumn we give double square so we can select 
#multiple columns as list

corr=tips[["total_bill","tip","size"]].corr()
corr
```

```python
from matplotlib.pyplot import annotate


sns.heatmap(corr,annot=True,cmap="coolwarm")
#annot To show value of co relation
```

#### Custom visualization

```python
import pandas


df=pandas.read_csv("sales_data_sample.csv")
df
```

```python
# In matplotlib we were aggregating and then creating plot but here 
# we have estimator argument which applies the aggregation on the fly
plt.figure(figsize=(10,8))
sns.barplot(x=df["Category"],y=df["Total"],estimator=sum)
```
