### Data visualization with Matplotlib
Matplotlib is a powerful library for python that enables the creation of static, animated and interactive visualizations. it is widely used for data visualization in data science and analytics. In this lesson we will cover the basics of matplotlib, including creating various types of plots and customizing them.

```python
import matplotlib.pyplot as plt
```

#### LineGraph

```python
x=[1,2,3,4,5]
y=[1,4,9,16,25]

plt.plot(x,y)
plt.xlabel("X axis")
plt.ylabel("y-axis")
```

#### Customized line plot

```python
 
from matplotlib import markers
from matplotlib.lines import lineStyles
from matplotlib.pyplot import grid


x=[1,2,3,4,5]
y=[1,4,9,16,25]

plt.plot(x,y,color='red',marker="*",linestyle="--",linewidth=2,markersize=5)
plt.grid(True)
plt.xlabel("X axis")
plt.ylabel("y-axis")
```

#### Multiplots
##### In 4 different plots


```python
# in four different plots

from turtle import color


x=[1,2,3,4,5]
y1=[1,4,9,16,25]
y2=[1,8,27,64,125]

# defines size of plot
plt.figure(figsize=(9,5))

# subplot creates number of sub plot 2*2 and 1 plot= 2,2,1
# 2,2 = 2*2= total 4 plots can be creates
plt.subplot(2,2,1)
plt.plot(x,y1,color='red')
plt.subplot(2,2,2)
plt.plot(y1,x,color='green')
plt.subplot(2,2,3)
plt.plot(x,y2,color='blue')
plt.subplot(2,2,4)
plt.plot(y2,x,color='yellow')
```

#### Bargraph


```python
categories=["A","B","C","D","E"]
values=[5,4,3,6,1]

plt.bar(categories,values,color='green')
```

#### Histogram
Histograms are used to represent the distribution of a dataset. They divide the data into bins and count the number of data points in each bin

```python
data=[1,2,3,1,2,4,1,2,5,2,5,2,5]
plt.hist(data,bins=5,color='red',edgecolor='black',rwidth=1,orientation="horizontal")
```

#### Scatterplot


```python
x=[1,2,3,4,5]
y=[3,4,6,5,3]
plt.scatter(x,y,color='blue',marker="x")
```

#### PieChart

```python
from matplotlib.pylab import exp
from sklearn.metrics import auc


labels=["A","B","C","D"]
sizes=[10,20,30,40]
color=['gold','blue','red','green']
explode=(0.9,0,0,0) # displace A as first tuple value is .2 rest are 0 so B C D remains intact


# autopct == to display percent %1.1f%%==> one decimal digit
plt.pie(sizes,explode=explode,labels=labels,colors=color,autopct="%1.1f%%",shadow=True)
```

#### Sales Data visualization

```python
import pandas as pd
sales_data=pd.read_csv("sales_data_sample.csv")
sales_data.head(5)
```

```python
total_sales_by_product=sales_data.groupby("Category").agg({"Total":"sum"})
#After groupby(), the resulting DataFrame has Category as its index, not a column  so total_sales_by_product["Category"] will give a KeyError
#To use Category in plt.bar(), reset the index:

total_sales_by_product=total_sales_by_product.reset_index()
```

```python
total_sales_by_product.plot(kind='bar',color='blue')
plt.bar(total_sales_by_product["Category"],total_sales_by_product["Total"])
```
