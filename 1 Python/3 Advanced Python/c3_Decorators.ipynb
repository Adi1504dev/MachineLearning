### Decorators

Decorators are powerful and flexible feature in python that allows you to modify the behaviour of a functions or a class method. They are commonly used to add funtionality to functions or methods without modifying there actual code . This lesson covers the basics of decorators, including how to create and use them

#### Function copy

```python
def welcome():
    print("Hello world")

wel=welcome
wel #Function welcome is copied to welcome and  will remain even if welcome function is deleted
wel()
del welcome
wel
wel()
```

#### Closures
Function inside functions

```python
def main_welcome():
    msg="welcome"
    def sub_welcome():
        print("Welcome to python corse")
        print(msg)# can use varible from main method
        print("learn this well")
    return sub_welcome()# Submethod returned 

main_welcome()
```

```python
def main_welcome(func,lst):
    msg="welcome"
    def sub_welcome():
        print("Welcome to python corse")
        print(func(lst))# can use varible from main method
        print("learn this well")
    return sub_welcome()# Submethod returned 

#We can pass methods as an argument to another method
#For eg rit now i will pass len in-build method as an argument
# We can use it inside sub method with variable name func

main_welcome(len,[1,3,2,3,4])
```

#### Decorators



```python
def main_welcome(func):
    msg="welcome"
    def sub_welcome():
        print("Welcome to python corse")
        func()
        print("learn this well")
    return sub_welcome()# Submethod returned 


def hello_python():
    print("hello Python")

main_welcome(hello_python)
```

```python
#Using Decorators 
def main_welcome(func):
    msg="welcome"
    def sub_welcome():
        print("Welcome to python corse")
        func()
        print("learn this well")
    return sub_welcome()# Submethod returned 



@main_welcome #Using @ we can decorate the method -- This method is same as main_welcome(hello_python)
def hello_python():
    print("hello Python")

hello_python # Just calling method
```

##### Example2
With Argument

```python
def repeat(n):
    def decorator(func):
        def wrapper(*args,**kwargs):

            for i in range(n):
                func(*args,**kwargs)
        return wrapper
    return decorator

@repeat(3)
def say_Hello(*args):
    print(f"Hello {args[0]} {args[1]}")
say_Hello("Aditya","Shobhita")

@repeat(3)
def say_Hi(t,b):#Tupple unpacking
    print(f"Hi {t} {b}")
say_Hi("Aditya","Shobhita") #Argument here "Aditya"---> directly goes to args value in sub function and from there it goes to say_hello
```

```python
from functools import wraps

def repeat(n):
    def decorator(func):
        @wraps(func)
        def wrapper(*args, **kwargs):

            
            start_num_base = args[0]
            remaining_args = args[1:]
            # Move the list INSIDE so it refreshes every time you call the function
            results_list = [] 
            for i in range(n):
                # Calculate the incremented input
                current_input = start_num_base + i
                
                # Execute the function
                value = func(current_input, *remaining_args, **kwargs)
                
                # 1. APPEND FIRST so the list isn't empty when you print it
                results_list.append(value)
                
                # 2. NOW PRINT
                print(f"Iteration {i+1}: Result was {value} and list so far is {results_list}")
            
            # Return the full list to the outside world
            return results_list
            
        return wrapper
    return decorator

@repeat(3)
def calculate_square(number):
    return number * number

my_data = calculate_square(5)

print("\n--- Final Results ---")
print(f"Final List: {my_data}")
```

```python
def repeat(n):
    def decorator(func):
        @wraps(func)
        def wrapper(*args, **kwargs):
            resultval=None
            for i in range(n):
                input=args[0]+i
                remaining_args = args[1:]
                resultval= func(input, *remaining_args, **kwargs)
            return resultval
        return wrapper
    return decorator
@repeat(3)
def calculate_square(number):
    return number * number
print(calculate_square(5))
```
