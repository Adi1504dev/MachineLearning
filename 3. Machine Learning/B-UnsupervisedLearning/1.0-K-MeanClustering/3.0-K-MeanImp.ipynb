```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
X,y=make_blobs(n_samples=1000,centers=3,n_features=2,random_state=42)
```

```python
X
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
```

```python
## Standadization- Feature scaling
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.33,random_state=42)
```

```python
scaler=StandardScaler()
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
from sklearn.cluster import KMeans
```

```python
## Elbow method to predict k value
wcss=[]
for k in range(1,11):
    kmeans=KMeans(n_clusters=k,init="k-means++")
    kmeans.fit(X_train)
    wcss.append(kmeans.inertia_)
```

```python
wcss
```

```python
## Plot Elbow curve
plt.plot(range(1,11),wcss)
plt.xticks(range(1,11))
plt.xlabel("Number of clusters")
plt.ylabel("k values")
# we see elbow at k=3
```

```python
 kmeans=KMeans(n_clusters=3,init="k-means++")
 kmeans.fit_predict(X_train)
```

```python
y_pred=kmeans.predict(X_test)
```

```python
y_pred
```

```python
plt.scatter(X_test[:,0],X_test[:,1],c=y_pred)
```

```python
## Silhoutee Scoring
from sklearn.metrics import silhouette_score
```

```python
silhouette_cofficient=[]
for k in range(2,11):
    kmean=KMeans(n_clusters=k,init='k-means++')
    kmean.fit(X_train)
    score=silhouette_score(X_train,kmean.labels_)
    silhouette_cofficient.append(score)

silhouette_cofficient
```

```python
## Plot Elbow curve
plt.plot(range(2,11),silhouette_cofficient)
plt.xticks(range(2,11))
plt.xlabel("silhouette_cofficient")
plt.ylabel("k values")
# we max score at k=3
```
