```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
```

```python
## Import Iris dataset
iris=datasets.load_iris()
```

```python
iris_data=pd.DataFrame(iris.data,columns=iris.feature_names)

X=iris_data
y=pd.DataFrame(iris.target,columns=['target'])
```

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
X=ss.fit_transform(X)
```

```python
#Apply pca
from sklearn.decomposition import PCA
pca=PCA(n_components=2)
pca_x=pca.fit_transform(X)
```

```python
plt.scatter(pca_x[:,0],pca_x[:,1],c=iris.target)
```

```python
## Agglomerative Clustering
## To construct Dendogram

import scipy.cluster.hierarchy as sc
plt.figure(figsize=(20,7))
plt.title("dendrograms")

## Create dendrogram
sc.dendrogram(sc.linkage(pca_x,method='ward'))
plt.xlabel("sample index")
plt.ylabel("distance")
```

```python
from sklearn.cluster import AgglomerativeClustering
'''Hierarchical Clustering has two components:

Metric (Affinity) → How distance is calculated.
Linkage → How distance between clusters is calculated.
'''

'''
Metric:
Distance(A,B)=5

Linkage:
Distance(Cluster1, Cluster2)
=
min/max/average/variance-based
of all point distances
'''

cluster=AgglomerativeClustering(n_clusters=2,linkage='ward')
cluster.fit(pca_x)
```

```python
cluster.labels_
```

```python
plt.scatter(pca_x[:,0],pca_x[:,1],c=cluster.labels_)
```
