### Bag of words

```python
import pandas as pd
df=pd.read_csv("data\SPAM_HAM_Data.txt")
```

```python
df
```

```python
import re
import nltk
```

```python
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
WordNetLemmatizer=WordNetLemmatizer()
```

```python
corpus=[]
stop_words = set(stopwords.words('english')) # do not craete it inside loop else will take too much time to run
for i in range(0,len(df)):
    review=re.sub("[^a-zA-Z]"," ",df['message'][i])
    review=review.split()
    review=[WordNetLemmatizer.lemmatize(word) for word in review if word not in stop_words]
    corpus.append(' '.join(review))

```

```python

# N-Grams
from sklearn.feature_extraction.text import TfidfVectorizer
#cv=CountVectorizer(max_features=2500) #Column=2500
tfIdf=TfidfVectorizer(max_features=50,ngram_range=(1,2)) #Column=50 #Keep up to 50 most frequent features.-- if unique words are less then it will be less
analyzer = tfIdf.build_analyzer()

X=tfIdf.fit_transform(corpus)
```
