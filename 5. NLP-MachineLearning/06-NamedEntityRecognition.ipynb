
### Named Entity Recognition
Named Entity Recognition (NER) is a Natural Language Processing (NLP) technique that identifies and extracts key information (entities) from unstructured text, classifying them into predefined categories like names of people, organizations, locations, dates, and monetary values

Person (PER): Names of individuals (e.g., Ada Lovelace, Elon Musk).

Organization (ORG): Companies, agencies, and institutions (e.g., Google, United Nations).

Location/Geopolitical (LOC/GPE): Countries, cities, and landmarks (e.g., Paris, Mount Everest).

Date & Time (DATE/TIME): Temporal expressions (e.g., October 2026, 3 p.m.).Other: 

Monetary values, percentages, medical codes, and product names

```python
paragraph = "Sundar Pichai is the CEO of Google."
```

```python
import nltk
```

```python
words=nltk.word_tokenize(paragraph)
```

```python
tagged_words=nltk.pos_tag(words)
```

```python
tagged_words
```

```python
#Required data
nltk.download('maxent_ne_chunker_tab')
nltk.download('words')
```

```python
#It takes tagged words and find named entitied out of it
entities=nltk.ne_chunk(tagged_tokens=tagged_words)
#enetities type=<class 'nltk.tree.Tree'>
print(entities)
```

```python
for chunk in entities:
    if hasattr(chunk, "label"):
        entity = " ".join(word for word, pos in chunk.leaves())
        print(entity, "->", chunk.label())
```

```python
#With Spacy
import spacy
nlp = spacy.load("en_core_web_sm")
text = "Sundar Pichai is very happy to be the CEO of Google."

doc = nlp(text)
for ent in doc.ents:
    print(ent.text, "->", ent.label_)
```
