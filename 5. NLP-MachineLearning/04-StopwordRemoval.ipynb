```python
corpus="""The students were studying different machine learning algorithms and analyzing large datasets. They were running experiments, comparing results, and making improvements to their models. Some researchers were working on predicting customer behavior, while others were focusing on optimizing business processes. The teams collaborated closely and shared their findings regularly to achieve better outcomes. Despite facing several challenges, they continued learning, adapting, and improving their techniques over time."""
```

```python
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords
```

```python
import nltk
nltk.download('stopwords')
```

```python
stopwords_value=stopwords.words('english')
```

```python
stemmer=PorterStemmer() # does not require english
sentences=nltk.sent_tokenize(corpus)
```

```python
sentences
```

```python
#Apply stopwords and filter and then apply stemming
for i in range(len(sentences)):
    words=nltk.word_tokenize(sentences[i])
    words=[stemmer.stem(word) for word in words if word.lower() not in set(stopwords.words('english'))]
    sentences[i]='|'.join(words)
#You can also apply snowballsetmmer and lemmitization
```

```python
# Resetting sentences
sentences=nltk.sent_tokenize(corpus)
```

```python
#Apply stopwords and filter and then apply snowball stemming
from nltk.stem import SnowballStemmer
SnowballStemmer=SnowballStemmer('english')
for i in range(len(sentences)):
    words=nltk.word_tokenize(sentences[i])
    words=[SnowballStemmer.stem(word) for word in words if word.lower() not in set(stopwords.words('english'))]
    sentences[i]=' '.join(words)
```

```python
sentences=nltk.sent_tokenize(corpus)
```

```python
from nltk.stem import WordNetLemmatizer
lim=WordNetLemmatizer()
for i in range(len(sentences)):
    words=nltk.word_tokenize(sentences[i])
    words=[lim.lemmatize(word,pos='v') for word in words if word.lower() not in set(stopwords.words('english'))]
    sentences[i]=' '.join(words)
```
