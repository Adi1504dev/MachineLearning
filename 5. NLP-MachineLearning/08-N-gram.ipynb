### Bag of words- Ngrams

```python
import pandas as pd
df=pd.read_csv("data\SPAM_HAM_Data.txt")
```

```python
df
```

```python
import re
import nltk
```

```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
corpus=[]
stop_words = set(stopwords.words('english')) # do not craete it inside loop else will take too much time to run
for i in range(0,len(df)):
    review=re.sub("[^a-zA-Z]"," ",df['message'][i])
    review=review.split()
    review=[ps.stem(word) for word in review if word not in stop_words]
    corpus.append(' '.join(review))

```

```python

# N-Grams
from sklearn.feature_extraction.text import CountVectorizer
#cv=CountVectorizer(max_features=2500) #Column=2500
cv=CountVectorizer(max_features=50,binary=True,ngram_range=(2,2)) #Column=50 #Keep up to 50 most frequent features.-- if unique words are less then it will be less
analyzer = cv.build_analyzer()

X=cv.fit_transform(corpus)
```

```python
cv.vocabulary_ #It give vocab and its index
'''
(1,1)--> unigram
1,2--> uni and bi
2,2--> bi
2,3--> bi and tri
'''
```

```python
from sklearn.feature_extraction.text import CountVectorizer

# Define your exact vocabulary with custom indices
my_vocab = {'machine': 0, 'learning': 1, 'nlp': 2}

# Initialize with the fixed vocabulary
cv = CountVectorizer(vocabulary=my_vocab) #fixed vocab
 #It give vocab and its index
X=cv.fit_transform(corpus)
cv.fixed_vocabulary_ #gives true in case of fixed vocab
```
