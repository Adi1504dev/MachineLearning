### Bag of words

```python
import pandas as pd
df=pd.read_csv("data\SPAM_HAM_Data.txt")
```

```python
df
```

```python
import re
import nltk
```

```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
corpus=[]
stop_words = set(stopwords.words('english')) # do not craete it inside loop else will take too much time to run
for i in range(0,len(df)):
    review=re.sub("[^a-zA-Z]"," ",df['message'][i])
    review=review.split()
    review=[ps.stem(word) for word in review if word not in stop_words]
    corpus.append(' '.join(review))

```

```python
## Create the bag of words model
## Using sklearn
## CountVectorize
'''
lowercase=True	Converts Machine → machine

stop_words='english'	Removes words like I, is, the

max_features=10	Keeps only top 10 most frequent features

min_df=1	Keep words appearing in ≥1 document

max_df=0.9	Remove words appearing in >90% of documents

ngram_range=(1,2)	Creates unigrams and bigrams

binary=False	Store actual counts
'''

'''
processing in cv
Raw Text
   ↓
lowercase
   ↓
tokenization
   ↓
stop word removal
   ↓
ngram creation
   ↓
vocabulary building
   ↓
count matrix
'''



'''
These method can be checked for the sentence
methods in cv
build_preprocessor()
    Raw text -> preprocessed text

build_tokenizer()
    Text -> list of tokens

build_analyzer()
    Full pipeline:
    preprocessing
    → tokenization
    → stopword removal
    → ngram generation

CountVectorizer internally uses the analyzer when creating features.

'''
from sklearn.feature_extraction.text import CountVectorizer
#cv=CountVectorizer(max_features=2500) #Column=2500
cv=CountVectorizer(max_features=500,binary=True) #Column=50 #Keep up to 50 most frequent features.-- if unique words are less then it will be less
analyzer = cv.build_analyzer()

print(analyzer(corpus[10]))
X=cv.fit_transform(corpus)
```

```python
X.toarray()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
y = le.fit_transform(df['label'])

print(y)
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20)
from sklearn.naive_bayes import MultinomialNB
spam_detect_model=MultinomialNB().fit(X_train,y_train)
y_pred=spam_detect_model.predict(X_test)
from sklearn.metrics import accuracy_score,classification_report
accuracy_score=accuracy_score(y_test,y_pred)
print(accuracy_score)
print(classification_report(y_test,y_pred))
```
