```python
import spacy
import nltk
```

```python
corpus="""
Hello Welcome, to Aditya's NLP tutorials.
Please do watch the entire course! to become expert in NLP.
"""
```

```python
print(corpus)
```

```python
## Tokenization
## Paragraph -->  Sentence
from nltk.tokenize import sent_tokenize
# U need punkt_tab for this 
## The Punkt tokenizer is a pre-trained, rule-based model that splits text into sentences:
nltk.download('punkt_tab')
#Full stop and ! are tretaed as aonther sentence
#This list of santences are called documents
documents=sent_tokenize(corpus)
```

```python
## Tokenization
## Paragraph -->  Sentence
from nltk.tokenize import word_tokenize
word_tokenize(corpus)
## ' ---> split words ,
#  !,.--> word 
# Aditya's --> Aditya , 's
```

```python
for sent in documents:
    print(word_tokenize(sent))
```

```python
from nltk.tokenize import wordpunct_tokenize
wordpunct_tokenize(corpus)
# diff ==> Aditya's --> Aditya, ' , s
```

```python
from nltk.tokenize import TreebankWordTokenizer
tokenizer=TreebankWordTokenizer()
tokenizer.tokenize(corpus)
# diff ==> only last full stop will be treated as sperate word
# rest all will be part of word only like tutorials.
```
